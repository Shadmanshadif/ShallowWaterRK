```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from the file
data = np.loadtxt('IntregralEuler1NoBordes.txt')

# Generate x values starting from 1
x = np.arange(1, len(data) + 1)

# Set the figure size
plt.figure(figsize=(15, 6))  # Adjust the width (15) and height (6) as needed

# Plot the data
plt.plot(x, data, marker='o')
plt.xlabel('Time')
plt.ylabel('Volume')
plt.title('Mass Conservation Euler1')

# Set the y-axis limits
plt.ylim(155, 165)

plt.show()


```
